```python
import pandas as pd

print("Loading sales data...")
# 1. Load the data (adjust the path if you are running this from a different folder)
df = pd.read_csv('processed_csv/sales_data_preprocessed.csv')

# 2. Ensure the Date column is a proper datetime object
df['Date'] = pd.to_datetime(df['Date']).dt.normalize()

# 3. Create a "mask" to filter only for October 1st to November 30th, 2025
start_date = pd.to_datetime('2025-10-01')
end_date = pd.to_datetime('2025-11-30')

mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
filtered_df = df.loc[mask]

# 4. Clean up non-product columns
if 'Time' in filtered_df.columns:
    filtered_df = filtered_df.drop(columns=['Time'])
filtered_df = filtered_df.drop(columns=['Date'])

# 5. Sum the sales for every product during this 2-month window
total_sales_per_product = filtered_df.sum(numeric_only=True)

# 6. Filter the list to only include products that sold more than 0 units
active_products = total_sales_per_product[total_sales_per_product > 0]

print("\n==================================================")
print(f"Products with sales between {start_date.date()} and {end_date.date()}")
print("==================================================")
print(f"Total active products: {len(active_products)}\n")

# Print the products sorted by their total sales volume (highest to lowest)
print(active_products.sort_values(ascending=False).to_string())


# Convert the product names into a clean Python list
product_list = active_products.index.tolist()

print(product_list)
```
